```python
# This notebook downloads and stores the OpenStreetMap NetworkX GraphML
# representation of the road network of Manhattan.
```

```python
import osmnx as ox
import matplotlib.pyplot as plt
import networkx as nx

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
CRS = 'EPSG:4326'
DATA_DIR = '../../data'
EXPORTS_DIR = f'{DATA_DIR}/exports'
```

```python
ny_mg = ox.graph_from_place('Manhattan, New York City, New York, USA')
```

```python
fig, ax = plt.subplots(figsize=(30, 30))
ox.plot_graph(ny_mg,
              ax=ax,
              bgcolor='#444',
              node_size=0,
              show=False,
              close=False,
              edge_color='#2c2c2c')
ax.set_facecolor('#444')
```

```python
ox.save_graphml(ny_mg, f'{EXPORTS_DIR}/graphml/manhattan/ny.graphml')
```
