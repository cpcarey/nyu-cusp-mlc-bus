```python
import geopandas as gpd
import numpy as npm
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from shapely.geometry import Point
from shapely.geometry import LineString
from shapely.geometry import MultiLineString

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
CRS_PLOT = 'EPSG:4326'
DATA_DIR = '../data'
EXPORTS_DIR = f'{DATA_DIR}/exports'

zips_gdf = gpd.read_file(f'{EXPORTS_DIR}/zips_bk.shp').to_crs({'init': CRS_PLOT})
lion_segments_gdf = gpd.read_file(f'{DATA_DIR}/lion_segments_bk.geojson').to_crs({'init': CRS_PLOT})
```

```python
lion_nodes_gdf = gpd.read_file(f'{DATA_DIR}/lion_nodes_bk.geojson').to_crs({'init': CRS_PLOT})
```

```python
fig, ax = plt.subplots(figsize=(15, 15))

zips_gdf.plot(ax=ax, color='lightgrey')
lion_segments_gdf.iloc[0:10000, :].plot(ax=ax, color='grey')

plt.show()
```

```python
for c in lion_segments_gdf.columns:
    print(c)
```

```python
lion_subset_gdf = lion_segments_gdf[['OBJECTID', 'NodeIDFrom', 'NodeIDTo']]
lion_subset_gdf = lion_subset_gdf.drop_duplicates(subset=['OBJECTID'])
lion_subset_gdf.head()
```

```python
len(lion_subset_gdf)
```

```python
edge_dict = {}
for i in range(0, len(lion_subset_gdf)):
    node_from = lion_subset_gdf.iloc[i, 1]
    node_to = lion_subset_gdf.iloc[i, 2]
    
    if node_from not in edge_dict:
        edge_dict[node_from] = set()
    edge_dict[node_from].add(node_to)
```

```python
edge_df = pd.DataFrame(columns=['node_from', 'nodes_to'])
for [key, value] in edge_dict.items():
    edge_df = edge_df.append({'node_from': key, 'nodes_to': value}, ignore_index=True)
edge_df.head()
```
