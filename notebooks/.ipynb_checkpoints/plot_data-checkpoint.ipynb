```python
import geopandas as gpd
import numpy as npm
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
CRS_PLOT = 'EPSG:4326'
DATA_DIR = '../data'

stops_gdf = gpd.read_file(f'{DATA_DIR}/NYC_2015_GTFS_Bus_Stops.shp').to_crs({'init': CRS_PLOT})
zips_gdf = gpd.read_file(f'{DATA_DIR}/ZIP_CODE_040114.shp').to_crs({'init': CRS_PLOT})
```

```python

def plot_stops(zips_gdf, stops_gdf):
    fig, ax = plt.subplots(figsize=(15, 15))
    zips_gdf.to_crs({'init': CRS_PLOT}).plot(ax=ax, color='lightgrey')
    stops_gdf.to_crs({'init': CRS_PLOT}).plot(ax=ax, color='black', markersize=2)
    plt.show()
```

```python
def filter_by_county(df, county):
    return df[df['COUNTY'] == county]

def join_to_zips(gdf):
    return gpd.sjoin(gdf, zips_gdf, how='inner', op='intersects')

brooklyn_stops_gdf = join_to_zips(stops_gdf)
brooklyn_stops_gdf = filter_by_county(brooklyn_stops_gdf, 'Kings')
brooklyn_zips_gdf = filter_by_county(zips_gdf, 'Kings')
tri_borough_zips_gdf = zips_gdf[zips_gdf['COUNTY'].isin(['Kings', 'Queens', 'New York'])]
quad_borough_zips_gdf = zips_gdf[zips_gdf['COUNTY'].isin(['Kings', 'Queens', 'New York', 'Richmond'])]
```

```python
brooklyn_stops_gdf.head()
```

```python
plot_stops(brooklyn_zips_gdf, brooklyn_stops_gdf)
```

```python
nyct_gdf = gpd.read_file(f'{DATA_DIR}/NYCT Bus Routes.shp').to_crs({'init': CRS_PLOT})
nyct_gdf = join_to_zips(nyct_gdf)
brooklyn_nyct_gdf = filter_by_county(nyct_gdf, 'Kings')
```

```python
plot_stops(quad_borough_zips_gdf, brooklyn_nyct_gdf)
```

```python
routes_df = pd.read_csv(f'{DATA_DIR}/routes.txt')
routes_df.head()
```

```python
brooklyn_nyct_gdf.columns
```

```python
def trim_gdf(gdf):
    columns = ['OBJECTID', 'route_id', 'route_color', 'geometry']
    return gdf[columns]

def clean_gdf(gdf):
    gdf['route_color'] = '#' + gdf['route_color']
    return gdf
    
brooklyn_nyct_gdf = brooklyn_nyct_gdf.merge(routes_df, on='route_id')
brooklyn_nyct_gdf = trim_gdf(brooklyn_nyct_gdf)
brooklyn_nyct_gdf = clean_gdf(brooklyn_nyct_gdf)
```

```python
def plot_routes(zips_gdf, routes_gdf):
    fig, ax = plt.subplots(figsize=(15, 15))
    zips_gdf.plot(ax=ax, color='lightgrey')
    routes_gdf.plot(ax=ax, color=routes_gdf['route_color'])
    plt.show()

plot_routes(quad_borough_zips_gdf, brooklyn_nyct_gdf)
```

```python
bus_times_single_df = bus_times_df[bus_times_df['inferred_route_id'] ==
                                   'MTA NYCT_B6']
bus_times_single_gdf = gpd.GeoDataFrame(bus_times_single_df,
                                        geometry=gpd.points_from_xy(
                                            bus_times_single_df['longitude'],
                                            bus_times_single_df['latitude'],
                                            crs=CRS_PLOT))
bus_times_single_gdf.head()
```

```python
def filter_lat_lons(df):
    return df[(df['longitude'] < -73.6) & (df['longitude'] > -74.3) & (df['latitude'] < 40.9) & (df['latitude'] > 40.4)]

def trim_bus_time(df):
    columns = ['latitude', 'longitude', 'vehicle_id', 'time_received', 'inferred_phase', 'inferred_trip_id', 'geometry', 'next_scheduled_stop_id']
    return df[columns]

def convert_time(df):
    df['time'] = pd.to_datetime(df['time_received'], format='%m/%d/%Y %H:%M', errors='coerce')
    columns = ['latitude', 'longitude', 'vehicle_id', 'time', 'inferred_phase', 'inferred_trip_id', 'geometry', 'next_scheduled_stop_id']
    return df[columns]
```

```python
bus_times_single_gdf = filter_lat_lons(bus_times_single_gdf)
bus_times_single_gdf = trim_bus_time(bus_times_single_gdf)
bus_times_single_gdf = convert_time(bus_times_single_gdf)
bus_times_single_gdf = bus_times_single_gdf.dropna()
bus_times_single_gdf.head()
```
